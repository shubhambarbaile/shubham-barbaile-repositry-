```python
#subplot() function is used to create a single subplot within a figure. 
#It takes three arguments that specify the number of rows, columns, and index of the subplot
import matplotlib.pyplot as plt
import numpy as np 
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(3,3,3)
plt.plot(x,y1)
plt.subplot(3,3,4)
plt.plot(x,y2)
plt.subplot(3,3,5)
plt.plot(x,y3)
plt.subplot(3,3,6)
plt.plot(x,y4)
```

```python
# The bar graph helps to compare the different sets of data among different groups easily. 
# It shows the relationship using two axes, in which the categories are on one axis and the discrete values are on the other axis

import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
#Box plots provide a quick visual summary of the variability of values in a dataset.
#They show the median, upper and lower quartiles, minimum and maximum values, and any outliers in thedataset


import matplotlib.pyplot as plt
import numpy as np 


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
cd = [box1,box2]
plt.boxplot(cd)
```
