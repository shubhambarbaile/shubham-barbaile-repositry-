```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Simple Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add a circle glyph
p.circle(x, y, size=10, color="navy", alpha=0.5)

# Show the plot in the notebook
output_notebook()
show(p)
```

```python
"""In Bokeh, glyphs are the visual shapes that represent your data on the plot, such as circles, lines, bars, patches, etc.
You can add glyphs to a Bokeh plot to display your data in various ways"""

from bokeh.plotting import figure, show

# Prepare data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Example Plot", x_axis_label='X-Axis', y_axis_label='Y-Axis')

# Add glyphs
# Example 1: Circle glyph
p.circle(x, y, size=10, color="navy", alpha=0.5)

# Example 2: Line glyph
p.line(x, y, line_width=2, color="green", legend_label="Line")

# Example 3: Square glyph
p.square(x, y, size=10, color="red", alpha=0.5)

# Example 4: Cross glyph
p.cross(x, y, size=10, color="orange", alpha=0.5)

# Example 5: Triangle glyph
p.triangle(x, y, size=10, color="purple", alpha=0.5)

# Customize plot
p.legend.location = "top_left"  # Move legend to top left corner

# Show the plot
show(p)
```
